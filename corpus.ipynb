```python
import nltk
from nltk.corpus import brown, inaugural, reuters, udhr

# Brown Corpus
print("Brown Corpus")
print("==============")
print("File IDs:")
print(brown.fileids()[:10])
print("Raw Text:")
print(brown.raw(fileids=brown.fileids()[0])[:50])
print("Words:")
print(brown.words()[:10])
print("Sentences:")
print(brown.sents()[:3])
print("Categories:")
print(brown.categories())
print("=" * 50)

# Inaugural Address Corpus
print("Inaugural Address Corpus")
print("========================")
print("File IDs:")
print(inaugural.fileids())
print("Raw Text:")
print(inaugural.raw('1789-Washington.txt')[:50])
print("Words:")
print(inaugural.words('1789-Washington.txt')[:10])
print("Sentences:")
print(inaugural.sents('1789-Washington.txt')[:3])
print("=" * 50)

# Reuters Corpus
print("Reuters Corpus")
print("==============")
print("File IDs:")
print(reuters.fileids()[:10])
print("Raw Text:")
print(reuters.raw(fileids=reuters.fileids()[0])[:50])
print("Words:")
print(reuters.words()[:10])
print("Sentences:")
print(reuters.sents()[:3])
print("Categories:")
print(reuters.categories())
print("=" * 50)

# UDHR Corpus
print("UDHR Corpus")
print("===========")
print("File IDs:")
print(udhr.fileids()[:10])
print("Raw Text:")
print(udhr.raw(fileids=udhr.fileids()[0])[:50])
print("Words:")
print(udhr.words()[:10])
print("Sentences:")
print(udhr.sents()[:3])
print("=" * 50)
```
